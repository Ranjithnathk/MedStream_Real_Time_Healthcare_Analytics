```python
storage = "healthcarestoragerk"
gold_container = "gold"

exports = [
  ("encounters_by_org_month", "powerbi/encounters_by_org_month"),
  ("encounters_by_department", "powerbi/encounters_by_department"),
  ("payer_coverage_summary", "powerbi/payer_coverage_summary"),
]

for src, dst in exports:
    src_path = f"abfss://{gold_container}@{storage}.dfs.core.windows.net/{src}"
    dst_path = f"abfss://{gold_container}@{storage}.dfs.core.windows.net/{dst}"

    df = spark.read.format("delta").load(src_path)
    df.write.mode("overwrite").parquet(dst_path)

print("Power BI Parquet exports completed")
```
